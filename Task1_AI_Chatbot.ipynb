```python
import nltk
import string
import numpy as np

from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
corpus = [
    "Hello",
    "Hi",
    "Good morning",
    "Artificial Intelligence is the simulation of human intelligence.",
    "Machine Learning is a subset of Artificial Intelligence.",
    "Deep Learning uses neural networks with multiple layers.",
    "Python is widely used for Artificial Intelligence.",
    "TensorFlow is a popular deep learning library.",
    "Scikit-learn is used for machine learning tasks.",
    "Goodbye"
]

responses = [
    "Hello! How can I help you?",
    "Hi there!",
    "Good morning!",
    "Artificial Intelligence allows machines to perform tasks that usually require human intelligence.",
    "Machine Learning enables systems to learn from data.",
    "Deep Learning uses neural networks for complex learning.",
    "Yes! Python is one of the best languages for AI.",
    "TensorFlow is developed by Google.",
    "Scikit-learn provides many ML algorithms.",
    "Goodbye! Have a great day."
]
```

```python
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    tokens = nltk.word_tokenize(text.lower())
    tokens = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word not in string.punctuation
    ]
    return " ".join(tokens)
```

```python
def chatbot(user_input):

    processed = preprocess(user_input)

    all_sentences = [preprocess(x) for x in corpus]

    vectorizer = TfidfVectorizer()

    vectors = vectorizer.fit_transform(all_sentences + [processed])

    similarity = cosine_similarity(vectors[-1], vectors[:-1])

    index = similarity.argmax()

    score = similarity[0][index]

    if score < 0.2:
        return "Sorry, I couldn't understand your question."

    return responses[index]
```

```python
print("🤖 AI Chatbot")
print("Type 'exit' to stop.\n")

while True:

    user = input("You : ")

    if user.lower() == "exit":
        print("Bot : Goodbye!")
        break

    print("Bot :", chatbot(user))
```
